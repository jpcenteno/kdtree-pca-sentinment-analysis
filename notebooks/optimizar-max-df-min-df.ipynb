La experimentaci√≥n sobre `max_df` y `min_df` fue llevada a cabo con el script `optimizar-max-df-min-df.py`. Esta notebook solo se encarga de graficar los resultados.

```python
import itertools
import time

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('results/resultados-opti-max-df-min-df.csv')
df.head()
```

```python
# Total elapsed time:
time_secs = df['time'].sum()
time.strftime('%H:%M:%S', time.gmtime(time_secs))
```

```python
# Grafica el resultado
def plot(df):
    df = df.drop_duplicates(['max_df', 'min_df'])
    acc = df.pivot('max_df', 'min_df', 'acc')
    plt.figure(figsize=(20,10))
    sns.heatmap(acc, annot=True, fmt="0.4f", linewidths=.5)
    
plot(df)
plt.savefig('plots/exp_maxmindf.png')
plt.show()
```
